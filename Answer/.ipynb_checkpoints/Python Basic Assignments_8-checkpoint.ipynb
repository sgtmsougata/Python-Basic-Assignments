```python
'''Q.1 Is the Python Standard Library included with PyInputPlus?'''

PyInputPlus is a Python module used for taking inputs with additional validation features.
PyInputPlus will keep asking the user for text until they enter valid input.

inputNum(), inputStr(), inputInt(), inputFloat(), inputBool(), inputChoice(), inputMenu(),
inputDate(), inputTime(), inputDatetime(), inputYesNo() and so on
```

```python
'''Q.2 Why is PyinputPlus commonly imported with import pyinputplus as pypi?'''

import pyinputplus as pyip so that we can enter a shorter name when calling the module's functions
```

```python
'''Q.3 How do you distinguish between inputInt() and inputFloat()?'''

inputInt() - Accept only integer value
inputFloat() - Accepts a floating-point numeric value
```

```python
'''Q.4 Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?'''

# pip install pyinputplus
import pyinputplus as pypi 

result = pypi.inputNum(prompt = 'Enter a whole number : ',
                       min = 0, max = 99, limit = 10)
print(result)
```

```python
'''Q.5 What is transferred to the keyword arguments allowRegexes and blockRegexes?'''

Regexes means regular expression, allowRegexes() accept this Regex whatever we input regex.
blockRegexes() block this Regex whatever we input regex.
```

```python
'''Q.6 If a blank input is entered three times, what does inputStr(limit=3) do?'''

It gives ValidationException error
pypi.inputStr(limit = 3)
```

```python
'''Q.7 If blank input is entered three times, what does inputStr(limit=3, default='hello') do?'''

After three times blank input it gives default value
pypi.inputStr(limit=3, default='hello')
```
