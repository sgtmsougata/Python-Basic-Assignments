```python
'''Q.1 What is the result of the code, and why?
def func(a, b=6, c=8):
    print(a, b, c)
func(1, 2)
'''
def func(a, b=6, c=8):
    print(a, b, c)
func(1, 2)

'''From this function a,b,c are the place holder and b default value 6 and c 8
at least one value require for exicute this func() function'''
```

```python
'''Q.2 What is the result of this code, and why?
def func(a, b, c=5):
    print(a, b, c)
func(1, c=3, b=2)
'''
def func(a, b, c=5):
    print(a, b, c)
func(1, c=3, b=2)

'''When we make function call, order will be positional argument and then keywords arguments. 
we can pass the keyword arguments in any order we want.'''
```

```python
'''Q.3 How about this code: what is its result, and why?
def func(a, *pargs):
    print(a, pargs)
func(1, 2, 3)
'''
def func(a, *pargs):
    print(a, pargs)
func(1, 2, 3)

'''The return type of *args parameter is tuple so pargs take (2,3)'''
```

```python
'''Q.4 What does this code print, and why?
def func(a, **kargs):
    print(a, kargs)
func(a=1, c=3, b=2)
'''
def func(a, **kargs):
    print(a, kargs)
func(a=1, c=3, b=2)

'''The return type of  **kargs is  dictionary'''
```

```python
'''Q.5 What gets printed by this, and explain?
def func(a, b, c=8, d=5): 
    print(a, b, c, d)
func(1, *(5, 6))
'''
def func(a, b, c=8, d=5): 
    print(a, b, c, d)
func(1, *(5, 6))

''' *' is the unpacking operator and are operators that unpack the values from iterable objects in Python. The single 
asterisk operator * can be used on any iterable that Python provides, while the double asterisk operator ** can only 
be used on dictionaries. In the example the value *(5,6) will be unpacked and will be assigned to b and c and passed 
as arguments, d =5 will taken by defaults are keyword arguments.
'''
```
